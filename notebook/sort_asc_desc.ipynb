```python
def is_sorted(li):
    asc = True
    desc = True
    for i in range(len(li)-1):
        if li[i] > li[i+1]:
            asc = False
            print('iteration: ', i, 'li[i]: ', li[i], 'li[i+1]: ', li[i+1], 'asc:', asc)
        elif li[i] < li[i+1]:
            desc = False
            print('iteration: ', i, 'li[i]: ', li[i], 'li[i+1]: ', li[i+1], 'desc:', desc)
    
    print('asc: ', asc, 'desc: ', desc)
    return asc or desc
```

```python
l1 = [1,2,3,4] # asc sorted
l2 = [1,2,3,2] # not sorted
l3 = [3,2,1,0] # desc sorted
l4 = [3,2,1,4] # not sorted
```

```python
import matplotlib.pyplot as plt
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, figsize=(12,7))

ax1.plot(l1, l1, c='r', marker='o')
ax1.set_title("l1")
ax2.plot(l1, l2, c='g', marker='o')
ax2.set_title("l2")
ax3.plot(l1, l3, c='b', marker='o')
ax3.set_title("l3")
ax4.plot(l1, l4, c='y', marker='o')
ax4.set_title("l4");
```
